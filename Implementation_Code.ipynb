## ticker list

```python
#get the stock ticker list from package
import yahoo_fin.stock_info as si
nasdaq_list = si.tickers_nasdaq() # ticker symbols for companies listed on the NASDAQ stock exchange.
other_list = si.tickers_other() #  list of ticker symbols for companies listed on other exchanges, such as the NYSE (New York Stock Exchange), AMEX (American Stock Exchange), and OTC (Over-the-Counter) markets
ETF_list = ['^DJI','^GSPC']
us_list = nasdaq_list + other_list + ETF_list
us_list = [x for x in us_list if str(x) != 'nan']
len(us_list)
```

## stock price for all

```python
import yfinance as yf
# yfinance is the Python package that sends API calls to Yahoo Finance, we use it to get historical price data
data1 = yf.download(tickers = us_list,
            period = "20y",
            interval = "1d",
            ignore_tz = True,
            prepost = False)
```

```python
data1 = data1.dropna(axis=1, how='all')
data1
```

## Import

```python
# We plan to use only adjusted close value for our analysis
adjclose_df = data1["Adj Close"]
adjclose_df
```

```python
tickerlist = adjclose_df.columns.tolist()
tickerlist
```

```python
# Connecting to MongoDB
from pymongo import MongoClient
client = MongoClient('localhost',27017)
db = client.apan5400
collection = db.project
```

```python
# Convert DataFrame to a list of dictionaries
adjclose_df_dict = adjclose_df.to_dict(orient='records')

# Insert data into MongoDB
for doc in adjclose_df_dict:
    doc_str_keys = {str(key): value for key, value in doc.items()}
    collection.insert_one(doc_str_keys)

print('Data imported to MongoDB successfully!')
```

```python
# Specify the nested keys as a list of strings in dot notation
tickerlist = adjclose_df.columns.tolist()

# Query documents and retrieve the specified fields
query = {}
projection = {field: 1 for field in tickerlist}  # Include all specified fields in the retrieved documents
documents = db.project.find(query, projection)

values = {field: [] for field in tickerlist}  # Create a dictionary to store the values for each field

for doc in documents:
    for field in tickerlist:
        value = doc.get(field)
        values[field].append(value)

df = pd.DataFrame(values)

# Display the DataFrame
print(df)
```

```python
#Calculate time series data
pct_df = df.pct_change()

#Generate correlation matrix
corr = pct_df.corr()
corr
```

```python
from pymongo import MongoClient
client = MongoClient('localhost',27017)
db = client.apan5400
collection = db.matrix

# Insert data into MongoDB
records = corr.to_dict(orient='records')
collection.insert_many(records)
```

```python
# Retrieve data from MongoDB
cursor = collection.find()
df = pd.DataFrame(list(cursor))
df.set_index('_id', inplace=True)
```

```python
# Using Dash to build interactive application
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.express as px


# Define the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Correlation Matrix"),
    html.Label("Select columns:"),
    dcc.Dropdown(
        id='dropdown',
        options=[{'label': col, 'value': col} for col in df.columns],
        value=[df.columns[0]],
        multi=True
    ),
    html.Div(id='correlation-matrix')
])

# Define the callback function that generates the correlation matrix
@app.callback(
    dash.dependencies.Output('correlation-matrix', 'children'),
    [dash.dependencies.Input('dropdown', 'value')])
def generate_correlation_matrix(selected_columns):
    if not selected_columns:
        return ""
    corr = df[selected_columns].corr()
    fig = px.imshow(corr)
    fig.update_xaxes(side="top")
    fig.update_layout(width=800, height=800)
    return dcc.Graph(figure=fig)

# Start the Dash app
if __name__ == '__main__':
    app.run_server(port=8095)
```
